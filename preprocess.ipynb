# Объединение в один датасет

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

directory = 'data'

df01_02 = pd.read_csv(directory + '/01_02.csv', encoding='windows-1251', sep=';')
df01_02['МесяцДатыЗаказа'] = df01_02['МесяцДаты Заказа']
df01_02.drop(['МесяцДаты Заказа'], axis=1, inplace=True)

df03_04 = pd.read_csv(directory + '/03_04.csv', encoding='windows-1251', sep=';')
df05_06 = pd.read_csv(directory + '/05_06.csv', encoding='windows-1251', sep=';')
df07_08 = pd.read_csv(directory + '/07_08.csv', encoding='windows-1251', sep=';')
df09_10 = pd.read_csv(directory + '/09_10.csv', encoding='windows-1251', sep=';')
df11 = pd.read_csv(directory + '/11.csv', encoding='windows-1251', sep=';')
df12 = pd.read_csv(directory + '/12.csv', encoding='windows-1251', sep=';')
```

```python
df12_columns = list(df12.columns[0].split('\t'))

def func(x):
    if x == '':
        return None
    else:
        return x

for i, column in tqdm(enumerate(df12_columns)):
    df12[column] = df12[df12.columns[0]].apply(lambda x: func(x.split('\t')[i]))

df12.drop([df12.columns[0]], axis=1, inplace=True) 
df12
```

```python

df01_02_2018 = pd.read_csv(directory + '/01_02_2018.csv', encoding='windows-1251', sep=';')
```

```python
dfs = [df01_02, df03_04, df05_06, df07_08, df09_10, df11, df12, df01_02_2018]
df = pd.concat(dfs, ignore_index=True)
df
```
