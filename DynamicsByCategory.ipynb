```python
df['Цена x Количество'] = df['Цена'] * df['Количество']
```

```python
from datetime import date

df['Дата_без_времени'] = df['Дата'].apply(lambda x: (str(x).split()[0]))

def get_iso_week_number(x):
    if (x == '01.01.2017'):
        return 0
    year = int(x.split('.')[2])
    month = int(x.split('.')[1])
    day = int(x.split('.')[0])
    d = date(year, month, day)

    adding = 0
    if (year == 2018):
        adding = date(2017, 12, 31).isocalendar()[1]
    return d.isocalendar()[1] + adding

df['ОтнНомерНедели'] = df['Дата'].apply(lambda x: get_iso_week_number(str(x).split()[0]))

def get_part(x):
    year = int(x.split('.')[2])
    month = int(x.split('.')[1])
    day = int(x.split('.')[0])
    if year == 2018:
        return 5
    if month <= 3:
        return 1
    elif month <= 6:
        return 2
    elif month <= 9:
        return 3
    else:
        return 4

df['Part'] = df['Дата'].apply(lambda x: get_part(str(x).split()[0]))

def get_weekday(x):
    year = int(x.split('.')[2])
    month = int(x.split('.')[1])
    day = int(x.split('.')[0])
    return date(year, month, day).isoweekday()

df['ДатаДеньНедели'] = df['Дата'].apply(lambda x: get_weekday(str(x).split()[0]))
df['ДатаДеньМесяца'] = df['Дата'].apply(lambda x: int(str(x).split()[0].split('.')[0]))

df['ДатаДеньНедели']
```

```python
from tqdm import tqdm

def count_statistics_monthly(df):
    sold_data = []
    placed_data = []
    months_names = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    for i in tqdm(range(1, 13)):
        placed = df[(df['МесяцДатыЗаказа'] == i) & (df['ГодДатыЗаказа'] == '2 017')]
        
        revenue = (placed['Количество'] * placed['Цена']).sum()
        major = placed['Маржа'].sum()
        count = placed['Количество'].sum()
        unique_orders = placed['НомерЗаказаНаСайте'].nunique()
        unique_clients = placed['Телефон_new'].nunique()
        average_order_revenue = placed[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = placed[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        placed_data.append(
            {
                'alias': f'{months_names[i - 1]} 2017',
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
        
        sold = df[(df['МесяцДатыЗаказа'] == i) & (df['ГодДатыЗаказа'] == '2 017') & (df['Статус'] == 'Доставлен') & (df['НовыйСтатус'] == 'Доставлен')]
    
        revenue = (sold['Количество'] * sold['Цена']).sum()
        major = sold['Маржа'].sum()
        count = sold['Количество'].sum()
        unique_orders = sold['НомерЗаказаНаСайте'].nunique()
        unique_clients = sold['Телефон_new'].nunique()
        average_order_revenue = sold[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = sold[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        sold_data.append(
            {
                'alias': f'{months_names[i - 1]} 2017',
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
        
    for i in tqdm(range(1, 3)):
        placed = df[(df['МесяцДатыЗаказа'] == i) & (df['ГодДатыЗаказа'] == '2 018')]
        
        revenue = (placed['Количество'] * placed['Цена']).sum()
        major = placed['Маржа'].sum()
        count = placed['Количество'].sum()
        unique_orders = placed['НомерЗаказаНаСайте'].nunique()
        unique_clients = placed['Телефон_new'].nunique()
        average_order_revenue = placed[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = placed[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        placed_data.append(
            {
                'alias': f'{months_names[i - 1]} 2018',
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
        
        sold = df[(df['МесяцДатыЗаказа'] == i) & (df['ГодДатыЗаказа'] == '2 018') & (df['Статус'] == 'Доставлен') & (df['НовыйСтатус'] == 'Доставлен')]
    
        revenue = (sold['Количество'] * sold['Цена']).sum()
        major = sold['Маржа'].sum()
        count = sold['Количество'].sum()
        unique_orders = sold['НомерЗаказаНаСайте'].nunique()
        unique_clients = sold['Телефон_new'].nunique()
        average_order_revenue = sold[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = sold[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        sold_data.append(
            {
                'alias': f'{months_names[i - 1]} 2018',
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
    return placed_data, sold_data

def count_statistics_weekly(df):
    sold_data = []
    placed_data = []
    months_names = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    for i in tqdm(range(1, 61)):
        placed = df[df['ОтнНомерНедели'] == i]
        
        month = placed['МесяцДатыЗаказа'].iloc[0]
        year = placed['ГодДатыЗаказа'].iloc[0]
        part = placed['Part'].iloc[0]
        
        revenue = (placed['Количество'] * placed['Цена']).sum()
        major = placed['Маржа'].sum()
        count = placed['Количество'].sum()
        unique_orders = placed['НомерЗаказаНаСайте'].nunique()
        unique_clients = placed['Телефон_new'].nunique()
        average_order_revenue = placed[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = placed[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        placed_data.append(
            {
                'alias': f'{months_names[month - 1]} {year} | {i}',
                'part': part,
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
        
        sold = df[(df['ОтнНомерНедели'] == i) & (df['Статус'] == 'Доставлен') & (df['НовыйСтатус'] == 'Доставлен')]
        if len(sold) == 0:
            sold_data.append(
                {
                    'alias': f'{months_names[month - 1]} {year} | {i}',
                    'part': part,
                    'revenue': 0,
                    'major': 0,
                    'count': 0,
                    'unique_orders': 0,
                    'unique_clients': 0,
                    'average_order_revenue': 0,
                    'major_per_product': 0,
                    'major_percent': 0,
                    'average_number_of_products': 0
                }
            )
            continue
             
        month = sold['МесяцДатыЗаказа'].iloc[0]
        year = sold['ГодДатыЗаказа'].iloc[0]
        part = sold['Part'].iloc[0]
    
        revenue = (sold['Количество'] * sold['Цена']).sum()
        major = sold['Маржа'].sum()
        count = sold['Количество'].sum()
        unique_orders = sold['НомерЗаказаНаСайте'].nunique()
        unique_clients = sold['Телефон_new'].nunique()
        average_order_revenue = sold[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = sold[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        sold_data.append(
            {
                'alias': f'{months_names[month - 1]} {year} | {i}',
                'part': part,
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
    return placed_data, sold_data


def count_statistics_daily(df):
    sold_data = []
    placed_data = []
    months_names = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    week_names = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вск']
    distinct_dates = list(map(lambda x: str(x), sorted(list(df['Дата_без_времени'].value_counts().index))))
    for date in tqdm(distinct_dates):
        placed = df[df['Дата_без_времени'] == date]
        year = placed['ГодДатыЗаказа'].iloc[0]
        month = placed['МесяцДатыЗаказа'].iloc[0]
        month_day = placed['ДатаДеньМесяца'].iloc[0]
        weekday = placed['ДатаДеньНедели'].iloc[0]        
        part = placed['Part'].iloc[0]
        
        revenue = (placed['Количество'] * placed['Цена']).sum()
        major = placed['Маржа'].sum()
        count = placed['Количество'].sum()
        unique_orders = placed['НомерЗаказаНаСайте'].nunique()
        unique_clients = placed['Телефон_new'].nunique()
        average_order_revenue = placed[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = placed[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        placed_data.append(
            {
                'alias': f'{month_day} {months_names[month - 1]} {year} | {week_names[weekday - 1]}',
                'part': part,
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
        sold = df[(df['Дата_без_времени'] == date) & (df['Статус'] == 'Доставлен') & (df['НовыйСтатус'] == 'Доставлен')]
        if len(sold) == 0:
            sold_data.append(
                {
                    'alias': f'{month_day} {months_names[month - 1]} {year} | {week_names[weekday - 1]}',
                    'part': part,
                    'revenue': 0,
                    'major': 0,
                    'count': 0,
                    'unique_orders': 0,
                    'unique_clients': 0,
                    'average_order_revenue': 0,
                    'major_per_product': 0,
                    'major_percent': 0,
                    'average_number_of_products': 0
                }
            )
            continue
        
        year = sold['ГодДатыЗаказа'].iloc[0]
        month = sold['МесяцДатыЗаказа'].iloc[0]
        month_day = sold['ДатаДеньМесяца'].iloc[0]
        weekday = sold['ДатаДеньНедели'].iloc[0]
        part = sold['Part'].iloc[0]
        
        revenue = (sold['Количество'] * sold['Цена']).sum()
        major = sold['Маржа'].sum()
        count = sold['Количество'].sum()
        unique_orders = sold['НомерЗаказаНаСайте'].nunique()
        unique_clients = sold['Телефон_new'].nunique()
        average_order_revenue = sold[['Цена x Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        major_per_product = major / count
        major_percent = 100.0 * major / revenue
        average_number_of_products = sold[['Количество', 'НомерЗаказаНаСайте']].groupby(['НомерЗаказаНаСайте']).sum().mean()
        sold_data.append(
            {
                'alias': f'{month_day} {months_names[month - 1]} {year} | {week_names[weekday - 1]}',
                'part': part,
                'revenue': round(float(revenue), 0),
                'major': round(float(major), 0),
                'count': round(int(count), 0),
                'unique_orders': unique_orders,
                'unique_clients': unique_clients,
                'average_order_revenue': round(float(average_order_revenue.iloc[0]), 1),
                'major_per_product': round(float(major_per_product), 1),
                'major_percent': round(float(major_percent), 1),
                'average_number_of_products': round(float(average_number_of_products.iloc[0]), 1)
            }
        )
    
    return placed_data, sold_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def make_stat_revenue(df, month, month_year, adding_title):
    sns.set_theme()
    
    placed_data_category_1_monthly, sold_data_category_1_monthly = count_statistics_monthly(df)
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    plt.title('Выручка в млн. рублей' + ' | ' + adding_title)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_monthly)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_monthly)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    merged['revenue_scaled'] = merged['revenue'] / 1000000.0
    sns.lineplot(data=merged, x='alias', y='revenue_scaled', hue='type')
    
    placed_data_category_1_weekly, sold_data_category_1_weekly = count_statistics_weekly(df)
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    plt.title('Выручка в млн. рублей' + ' | ' + adding_title)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_weekly)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_weekly)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    merged['revenue_scaled'] = merged['revenue'] / 1000000.0
    sns.lineplot(data=merged, x='alias', y='revenue_scaled', hue='type')

    placed_data_category_1_daily, sold_data_category_1_daily = count_statistics_daily(df[(df['МесяцДатыЗаказа'] == month) & (df['ГодДатыЗаказа'] == month_year)])
    months_names = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    plt.title(f'Выручка в тыс. рублей {month_year} | {months_names[month - 1]}' + ' | ' + adding_title)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_daily)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_daily)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    merged['revenue_scaled'] = merged['revenue'] / 1000.0
    sns.lineplot(data=merged, x='alias', y='revenue_scaled', hue='type')

def make_stat_count(df, month, month_year, adding_title):
    sns.set_theme()
    
    placed_data_category_1_monthly, sold_data_category_1_monthly = count_statistics_monthly(df)
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_monthly)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_monthly)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    merged['count_scaled'] = merged['count'] / 1000.0
    sns.lineplot(data=merged, x='alias', y='count_scaled', hue='type')
    plt.title('Кол-во товаров в тыс.' + ' | ' + adding_title)
    
    placed_data_category_1_weekly, sold_data_category_1_weekly = count_statistics_weekly(df)
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_weekly)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_weekly)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    merged['count_scaled'] = merged['count'] / 1000.0
    sns.lineplot(data=merged, x='alias', y='count_scaled', hue='type')
    plt.title('Кол-во товаров в тыс.' + ' | ' + adding_title)

    placed_data_category_1_daily, sold_data_category_1_daily = count_statistics_daily(df[(df['МесяцДатыЗаказа'] == month) & (df['ГодДатыЗаказа'] == month_year)])
    months_names = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    fig, ax = plt.subplots(figsize=(14, 5))
    plt.xticks(rotation=70)
    placed_df = pd.DataFrame.from_dict(placed_data_category_1_daily)
    placed_df['type'] = 'Размещенные'
    sold_df = pd.DataFrame.from_dict(sold_data_category_1_daily)
    sold_df['type'] = 'Выкупленные'
    merged = pd.concat([placed_df, sold_df])
    sns.lineplot(data=merged, x='alias', y='count', hue='type')
    plt.title(f'Кол-во товаров {month_year} | {months_names[month - 1]}' + ' | ' + adding_title)
```
