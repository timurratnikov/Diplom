```python
import pandas as pd
import numpy as np

directory = 'data'

df = pd.read_csv(directory + '/cleared_dataset.csv')
df
```

```python
df.info()
```

```python
df[['Тип', 'ТипТовара']]
```

```python
df['Тип'].value_counts()
```

```python
stat = pd.DataFrame()
stat['percent'] = (df['Группа2'].value_counts(normalize=True) * 100).apply(lambda x: round(x, 2))
stat['count'] = df['Группа2'].value_counts()
stat['count_per_month'] = (stat['count'] / 14.0).apply(lambda x: round(x))
stat['count_per_day'] = (stat['count'] / (14.0 * 30.0)).apply(lambda x: round(x))
stat
```

```python
df1 = df[(df['Статус'] == 'Доставлен') & (df['НовыйСтатус'] == 'Доставлен')]
stat = pd.DataFrame()
stat['percent'] = (df1['Группа2'].value_counts(normalize=True) * 100).apply(lambda x: round(x, 2))
stat['count'] = df1['Группа2'].value_counts()
stat['count_per_month'] = (stat['count'] / 14.0).apply(lambda x: round(x))
stat['count_per_day'] = (stat['count'] / (14.0 * 30.0)).apply(lambda x: round(x))
stat
```

```python
df['count'] = 1;
grouped = df[['count', 'ГодДатыЗаказа', 'МесяцДатыЗаказа']].groupby(['ГодДатыЗаказа', 'МесяцДатыЗаказа']).sum()
grouped['percent'] = (grouped['count'] / len(df) * 100.0).apply(lambda x: round(x, 2))
grouped
```
